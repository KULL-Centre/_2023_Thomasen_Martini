# Dimerization free energy of transmembrane domains

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.integrate as sint

def read_data(files):
    data = []
    for file in files:
        df = pd.read_csv(file, sep="\s+", skiprows=17, header=None)
        data.append(df[1])
    return data

def calculate_pmf(data):
    density_diff = data - data.iloc[-1]
    pmf = np.mean(density_diff, axis=1)
    std = np.std(density_diff, axis=1)
    return pmf, std
```

```python
methods = ["Unmodified", "PW", "PP"]
proteins = ["EphA1", "ErbB1"]
labels = ["Unmodified", "$\u03BB_{PW}$ = 1.10", "$\u03BB_{PP}$ = 0.88"]
colors = ['#1F78B4', '#FF7F00', '#33A02C'] 
ecolors = ['lightblue', (1.0, 0.8, 0.5), 'lightgreen'] 

fig, axes = plt.subplots(1, len(proteins), figsize=(15, 5), sharex=True, sharey=True)
fig.text(0.5, 0.001, 'Lateral distance of peptide COMs (nm)', ha='center', va='top', fontsize=18)
fig.text(0.0001, 0.5, 'G(r) (kJ/mol)', ha='center', va='center', fontsize=18, rotation=90)


minima_values = []
minima_errors = []

for protein, ax in zip(proteins, axes):
    r = []
    pmf_all = []

    for method in methods:
        pmf_method = []

        for i in range(2, 6):
            data = pd.read_csv(f"{protein}/{method}/profile_2000_{i}.xvg", sep="\s+", skiprows=17, header=None)
            reac_coord = data[0]
            data1 = data[1]
            pmf = data1 - data1.iloc[-1]
            pmf_method.append(pmf)
            if i == 2 and method == "Unmodified":
                r.append(reac_coord)

        pmf_all.append(pmf_method)

    pmf_all = np.array(pmf_all)
    pmf_mean = np.mean(pmf_all, axis=1)
    pmf_std = np.std(pmf_all, axis=1)

    for i, method in enumerate(methods):
        ax.errorbar(r[0], pmf_mean[i], yerr=pmf_std[i], label=labels[i], fmt='-', color=colors[i], ecolor=ecolors[i], capsize=5, elinewidth=1, alpha=1)
       
        if method != "PP":
            minima_index = np.argmin(pmf_mean[i])
            minima_value = pmf_mean[i][minima_index]
            minima_error = pmf_std[i][minima_index]
            minima_values.append(minima_value)
            minima_errors.append(minima_error)
            print(f"{protein} - {method}: Minimum Value = {minima_value}, Error = {minima_error}")


    ax.set_title(f"Protein: {protein}")
    ax.legend(fontsize=12)


    ax.axhline(y=0, color='black', linestyle='--', linewidth=1)
    ax.set_xticks(np.arange(0.6, 3.6, 0.4))
    ax.set_xlim([0.6, 3.4])
    
    for spine in ['bottom', 'top', 'left', 'right']:
        ax.spines[spine].set_color('black')
        ax.spines[spine].set_linewidth(0.7)
    
    ax.tick_params(axis='x', colors='black', labelsize=15, direction='out', length=3, width=0.7)
    ax.tick_params(axis='y', colors='black', labelsize=15, direction='out', length=3, width=0.7)

    
axes[0].set_title('EphA1', fontweight='bold', fontsize=16)
axes[1].set_title('ErbB1', fontweight='bold', fontsize=16)

axes[0].axhline(y=-15.4, color='red', linestyle='--', linewidth=1)
axes[1].axhline(y=-10.5, color='red', linestyle='--', linewidth=1)

plt.tight_layout()
plt.savefig("TM_pmf_profile.pdf", bbox_inches='tight')
plt.savefig("TM_pmf_profile.png", bbox_inches='tight', dpi=900)
plt.show()
```
